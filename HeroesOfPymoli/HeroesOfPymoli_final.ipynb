### Heroes Of Pymoli Data Analysis
* Of the 1163 active players, the vast majority are male (84%). There also exists, a smaller, but notable proportion of female players (14%).

* Our peak age demographic falls between 20-24 (44.8%) with secondary groups falling between 15-19 (18.60%) and 25-29 (13.4%).

```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file)
```

## Player Count

* Display the total number of players


```python
# Grab the total number of unique screen names
total_players = purchase_data['SN'].nunique()

# Build DataFrame
players_df = pd.DataFrame({
    "Total Players" : [total_players]
    })

players_df.head()
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Formats the output for the two columns that deal with monetary values
pd.options.display.float_format = '${:,.2f}'.format

# Build DataFrame
total_purchases_df = pd.DataFrame({
    '# of Unique Items' : [purchase_data['Item Name'].nunique()],
    'Average Item Price' : [purchase_data['Price'].mean()],
    'Toal Purhcases' : [purchase_data['Purchase ID'].nunique()],
    'Total Revenue' : [purchase_data['Price'].sum()]
})

total_purchases_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Info and calculations for DataFrame
gender_shape = purchase_data.groupby(['Gender'])
gender_counts = gender_shape.nunique()['SN']
percentage_of_players = gender_counts / total_players * 100

# Build DataFrame
gender_demographics_df = pd.DataFrame({
    'Total Counts' : gender_counts,
    'Percentage of Players' : percentage_of_players
})

# Optional: give the displayed data cleaner formatting
pd.options.display.float_format = '{:,.2f}'.format
gender_demographics_df.index.name = None

gender_demographics_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Info and calculations for DataFrame
purchase_counts = gender_shape['Purchase ID'].count()
average_purchase_price = gender_shape['Price'].mean()
total_purhcase_value = gender_shape['Price'].sum()
average_purhcase_person = total_purhcase_value / gender_counts

# Build DataFrame
purchase_analysis_gender_df = pd.DataFrame({
    'Purchase Count' : purchase_counts,
    'Average Purhcase Price' : average_purchase_price,
    'Total Purhcase Value' : total_purhcase_value,
    'Average Total Purhcase Per Person' : average_purhcase_person
})

# Optional: give the displayed data cleaner formatting
pd.options.display.float_format = '${:,.2f}'.format

purchase_analysis_gender_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establish bins for ages
age_bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 9999]
age_labels = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

# Segment and sort age values into bins established above
purchase_data['Age'] = pd.cut(purchase_data['Age'], age_bins, labels = age_labels)

# Info and calculations for DataFrame
age_demographics_grouped = purchase_data.groupby('Age')
purchase_count_age = age_demographics_grouped['SN'].nunique()
percentage_of_players_age = purchase_count_age / total_players * 100

# Build DataFrame
age_demographics_df = pd.DataFrame({
    'Percentage of Players' : percentage_of_players_age, 'Total Count' : purchase_count_age})

# Optional: give the displayed data cleaner formatting
age_demographics_df.index.name = None
age_demographics_df.style.format({"Percentage of Players":"{:,.2f}"})
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Info and calculations for DataFrame
purchase_count_analysis = age_demographics_grouped['Purchase ID'].count()
average_analysis = age_demographics_grouped['Price'].mean()
total_purchase_value_analysis = age_demographics_grouped['Price'].sum()
average_purchase_analysis = total_purchase_value_analysis / purchase_count_age

# Build DataFrame
purchase_analysis_age_df = pd.DataFrame({
    'Purchase Count' : purchase_count_analysis,
    'Average Purchase Price' : average_analysis,
    'Total Purchase Value' : total_purchase_value_analysis,
    'Average Total Purchase per Person' : average_purchase_analysis
})

# Optional: give the displayed data cleaner formatting
purchase_analysis_age_df.index.name = None

# Display the summary data frame
purchase_analysis_age_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Info and calculations for DataFrame
spending_info = purchase_data.groupby('SN')
spenders_purhcases = spending_info['Purchase ID'].count()
spenders_average_spent = spending_info['Price'].mean()
spenders_total_value = spending_info['Price'].sum()

# Build DataFrame
top_spenders_df = pd.DataFrame({
    'Purchase Count' : spenders_purhcases,
    'Average Purchase Price' : spenders_average_spent,
    'Total Purchase Value': spenders_total_value,
})

# Optional: give the displayed data cleaner formatting
top_spenders_df.index.name = None

# Sort by highest purchase value
ranked_spenders = top_spenders_df.sort_values(['Total Purchase Value'], ascending = False)

ranked_spenders.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Info and calculations for DataFrame
most_popular_items = purchase_data[['Item ID', 'Item Name', 'Price']]
most_popular_items_grouped = most_popular_items.groupby(['Item ID', 'Item Name'])
purchased_item_count = most_popular_items_grouped['Price'].count()
purchased_item_value = (most_popular_items_grouped['Price'].sum())
purchased_item_price = purchased_item_value / purchased_item_count

# Build DataFrame
most_popular_df = pd.DataFrame({
    'Purchase Count' : purchased_item_count,
    'Item Price' : purchased_item_price,
    'Total Purchase Value' : purchased_item_value
})

# Sort by highest purchase value
ranked_popular_items = most_popular_df.sort_values(['Purchase Count'], ascending = False)

ranked_popular_items.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
# Sort by highest purchase value
ranked_popular_items = most_popular_df.sort_values(['Total Purchase Value'], ascending = False)

ranked_popular_items.head()
```
